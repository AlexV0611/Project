```python
df_attrib = df_attrib.dropna()
```

```python
# Buscar duplicados
duplicados = df_attrib.duplicated()

# Imprimir las filas duplicadas
print("Filas duplicadas:")
print(df_attrib[duplicados])
```

```python
canciones_por_artista = df_attrib.groupby('artistName')['trackName'].apply(list)

# Imprimir las canciones por artista
for artista, canciones in canciones_por_artista.items():
    print(f"{artista}: {', '.join(canciones)}")
```

```python
# Identificar canciones duplicadas
duplicados = df_attrib[df_attrib.duplicated(subset=['trackName', 'artistName'], keep=False)]

# Imprimir las canciones duplicadas
print("Canciones Duplicadas:")
print(duplicados)
```

```python
# Contar las canciones repetidas
conteo_canciones_repetidas = duplicados.groupby(['trackName', 'artistName']).size().reset_index(name='Count')

# Imprimir el conteo de canciones repetidas
print("Conteo de Canciones Repetidas:")
print(conteo_canciones_repetidas)
```

```python
df_attrib_final = df_attrib.drop(['trackName', 'artistName', 'msPlayed', 'duration_ms', 'type',
                                  'id', 'uri', 'track_href', 'analysis_url',
                                  'duration_ms', 'time_signature'], axis=1)

df_attrib_final


```

```python
df_attrib_final= df_attrib_final.reset_index(drop=True)
```

```python
# Contar el número de géneros únicos
num_generos_unicos = df_attrib_final['genre'].nunique()

# Mostrar el resultado
print(f"Número de géneros únicos: {num_generos_unicos}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder


# Separar las características y la variable objetivo
X = df_attrib_final.drop('genre', axis=1)
y = df_attrib_final['genre']

# Codificar la variable objetivo
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Escalar las características usando RobustScaler
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo SVM
svm_clf = SVC(kernel='linear', C=1.0, random_state=42)
svm_clf.fit(X_train_scaled, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = svm_clf.predict(X_test_scaled)

# Imprimir el informe de clasificación
print(classification_report(y_test, y_pred))
```

MLP

```python
from sklearn.neural_network import MLPClassifier

# Crear y entrenar el modelo MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
mlp_clf.fit(X_train_scaled, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = mlp_clf.predict(X_test_scaled)

# Imprimir el informe de clasificación
print(classification_report(y_test, y_pred))
```

Modificamos hiperparámetros

```python
# Definir el espacio de búsqueda de hiperparámetros
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (150,)],
    'max_iter': [200, 300, 500],
    'alpha': [0.0001, 0.001, 0.01],
}
```

```python
from sklearn.model_selection import GridSearchCV

#Configurar GridSearchCV
grid_search = GridSearchCV(mlp_clf, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
```

```python
# Realizar la búsqueda de hiperparámetros
grid_search.fit(X_train_scaled, y_train)
```

```python
# Obtener los mejores parámetros
best_params = grid_search.best_params_
print("Mejores parámetros:", best_params)


```

```python
# Obtener el mejor modelo
best_model = grid_search.best_estimator_
```

```python
# Realizar predicciones en el conjunto de prueba con el mejor modelo
y_pred = best_model.predict(X_test_scaled)

# Imprimir el informe de clasificación con el mejor modelo
print("Informe de clasificación con mejores parámetros:")
print(classification_report(y_test, y_pred))
```

Validación cruzada:

```python
from sklearn.model_selection import cross_val_score

# Realizar validación cruzada
cv_scores = cross_val_score(best_model, X_train_scaled, y_train, cv=5, scoring='accuracy')

# Imprimir los resultados de la validación cruzada
print("Resultados de la Validación Cruzada:")
print("Scores por cada fold:", cv_scores)
print("Accuracy promedio:", np.mean(cv_scores))
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train_scaled, y_train)
```
